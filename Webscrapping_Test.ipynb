```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

# Question 1: Use yfinance to Extract Tesla Stock Data

```python
# Create a Ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Extract historical stock data and save it in a dataframe called tesla_data
tesla_data = tesla.history(period="max")

# Reset the index, save, and display the first five rows of the tesla_data dataframe
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

# Question 5: Plot Tesla Stock Graph

```python
def make_graph(stock_data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(stock_data['Date'], stock_data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price USD')
    plt.show()

# Use the make_graph function to graph the Tesla Stock Data with a title
make_graph(tesla_data, "Tesla Stock Price Over Time")
```
